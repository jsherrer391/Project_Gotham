```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
#Mock data
daily_returns= pd.DataFrame({
    "SPY":[0.5, 0.65, 0.62],
    "IWM":[0.45, 0.63, 0.55],
    "AGG":[0.99, 0.87, 0.96]})
daily_returns
```

This function evaluates the risk profile of each ETFs by calculating the standard deviation and sorting it from the smallest to the largest one.

```python
def calc_and_sort_standard_deviation(daily_returns):
    standard_deviation_sorted= daily_returns.std().sort_values()
    print(
          f"The *least* volatile ETF is {standard_deviation_sorted.index[0]} with a Standard Deviation of: {round(standard_deviation_sorted[0], 2)}\n"
          f"The *most* volatile ETF is {standard_deviation_sorted.index[-1]} with a Standard Deviation of: {round(standard_deviation_sorted[-1], 2)}\n"
    )
    return standard_deviation_sorted
```

```python
def calc_annual_standard_deviation(std):
    annual_standard_deviation = std * np.sqrt(252)
    print(
          f"When annualized, the *least* volatile ETF is {annual_standard_deviation.index[0]} with a STD of: {round(annual_standard_deviation[0], 2)}\n"
          f"When annualized, the *most* volatile ETF is {annual_standard_deviation.index[-1]} with a STD of: {round(annual_standard_deviation[-1], 2)}\n"
    )
    return annual_standard_deviation
```

```python
sorted_std = calc_and_sort_standard_deviation(daily_returns)
annual_sorted_std = calc_annual_standard_deviation(sorted_std)
```

```python
def calc_rolling_std_and_plot(daily_returns, x):
    rolling_std = daily_returns.rolling(window = x).std()
    rolling_std.plot(figsize= (20,10), 
                 title = '21-day rolling standard deviation for ETFs ')

calc_rolling_std_and_plot(daily_returns, 2)
```
